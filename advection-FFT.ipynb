```python
dt = (netCDF4.num2date(grid1.axes['time']['data'][0],grid1.axes['time']['units']) - \
                netCDF4.num2date(grid2.axes['time']['data'][0],grid2.axes['time']['units'])).seconds
```

```python
t1 = netCDF4.num2date(grid1.axes['time']['data'][0],grid1.axes['time']['units'])
t2 = netCDF4.num2date(grid2.axes['time']['data'][0],grid2.axes['time']['units'])
dt = (t2 - t1).seconds

dx = grid1.axes['x_disp']['data'][1] - grid1.axes['x_disp']['data'][0]
dy = grid1.axes['y_disp']['data'][1] - grid1.axes['y_disp']['data'][0]
print dx, dy
```

```python
image1 = grid2.fields['rain_z']['data'][0,:,:]
print image1.shape
image2 = grid1.fields['rain_z']['data'][0,:,:]
```

```python
# discrete fast fourier transformation and complex conjugation of image 2
#
image1FFT = np.fft.fft2(image1)
image2FFT = np.conjugate( np.fft.fft2(image2) )
```

```python
# inverse fourier transformation of product -> equal to cross correlation
#
imageCCor = np.real( np.fft.ifft2( (image1FFT*image2FFT) ) )
```

```python
# Shift the zero-frequency component to the center of the spectrum
#
imageCCorShift = np.fft.fftshift(imageCCor)
```

```python
# determine the distance of the maximum from the center
#
row, col = image1.shape

yShift, xShift = np.unravel_index( np.argmax(imageCCorShift), (row,col) )

yShift -= int(row/2)
xShift -= int(col/2)

print("shift of image1 in x-direction [pixel]: " + str(xShift))
print("shift of image1 in y-direction [pixel]: " + str(yShift))
```

```python
x_movement = xShift * dx
y_movement = yShift * dy
u = x_movement/dt
v = y_movement/dt
print x_movement, y_movement, u, v
```

```python
x = grid1.axes['x_disp']['data']
y = grid1.axes['y_disp']['data']
fig = plt.figure(figsize = [15,10])
plt.pcolormesh(x,y,image1)
plt.contour(x+x_movement, y+y_movement, image2)
```
