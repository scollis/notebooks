```python
infile = '/Users/scollis/Library/CloudStorage/Box-Box/Kestrel/ENVIRO_-_2851262_Apr_16_2024_3_24_24_PM.csv'
data = pandas.read_csv(infile, header=3)
units = data.iloc[0]
data = data[1:]
data.Temperature = data.Temperature.astype(float)
data['Dew Point'] = data['Dew Point'].astype(float)
data['Wind Speed'] = data['Wind Speed'].astype(float)
data['time'] = pandas.to_datetime(data['FORMATTED DATE_TIME'])
data = data.set_index('time')
```

```python
fm = plt.figure(figsize=[10,8])
data['Dew Point'].plot(color='blue', label='Dewpoint(c)')
data['Temperature'].plot(color='red', label='Temperature(c)')
ax1 = plt.gca()
ax1.set_ylabel('Temperature (c)')


plt.legend()
ax2 = plt.twinx()
data['Wind Speed'].plot(ax=ax2, color='green')
ax2.set_ylabel('Speed (km/h)')

```

```python
fm = plt.figure(figsize=[10,8])
((data['Temperature'] * 9./5.) + 32.).plot(color='red', label='Temperature(f)')
((data['Dew Point'] * 9./5.) + 32.).plot(color='blue', label='Dew Point(f)')
ax1 = plt.gca()
ax1.set_ylabel('Temperature (f)')


plt.legend()
ax2 = plt.twinx()
data['Wind Speed'].plot(ax=ax2, color='green')
ax2.set_ylabel('Speed (km/h)')

```
