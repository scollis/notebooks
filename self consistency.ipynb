```python
my_display = pyart.graph.RadarDisplay(myradar)
tilt = 1

f = figure(figsize=[15,5])
subplot(1,2,1)
my_display.plot_ppi('reflectivity_horizontal', tilt, vmin=-6, vmax=64.0)
gca().set_title(my_display.generate_title('reflectivity_horizontal', tilt))
subplot(1,2,2)
my_display.plot_ppi('recalculated_diff_phase', tilt, vmin=-1, vmax=6)#, vmin=-nyq, vmax=nyq)
gca().set_title(my_display.generate_title('recalculated_diff_phase', tilt))
```

```python
good = np.where(myradar.fields['norm_coherent_power']['data'][:,0:-20].flatten() > .5)[0] 
H, xedges, yedges = np.histogram2d(my_display.z[:,0:-20].flatten()[good]/1000.0,
    myradar.fields['corrected_reflectivity_horizontal']['data'][:,0:-20].flatten()[good],
     bins=(62, 60), range=([0,17], [-10,64]))
```

```python
fig = plt.figure(figsize=[8,5])
plt.pcolormesh(yedges, xedges, H)
plt.xlabel('Refl fac (dBZ)'); plt.ylabel('height (m)');cb=plt.colorbar()
cb.set_label('ngates')
plt.title('DBZ')
```

```python
coherent = myradar.fields['norm_coherent_power']['data'][:,0:-20].flatten() > .5
water = my_display.z[:,0:-20].flatten()/1000.0 < 3.0
good_and_liquid = np.where(np.logical_and(coherent, water))[0]
refl_range = [-10,64]
kdp_range = [-1,9]

sc_H, sc_xedges, sc_yedges = np.histogram2d(myradar.fields['corrected_reflectivity_horizontal']['data'][:,0:-20].flatten()[good_and_liquid],
    myradar.fields['recalculated_diff_phase']['data'][:,0:-20].flatten()[good_and_liquid],
     bins=(100, 100), range=(refl_range, kdp_range))
```

```python

fig = plt.figure(figsize=[12,7])
pcm = plt.pcolormesh(sc_yedges, sc_xedges, np.log10(sc_H+0.01), vmin=0.0)
plt.xlabel('Specific differential Phase ($\mathrm{\circ{}/km}$)')
plt.ylabel('Eq. Reflectivity Factor (dBZ)');cb=plt.colorbar()
gca().axes.set_ylim(refl_range)
gca().axes.set_xlim(kdp_range)
cb.set_label(r'log$_{10}$($\mathrm{n_{gates})}$')
title=plt.title('Self consistency test of $\mathrm{Z_e}$ and $\mathrm{K_{DP}}$ ')
title.set_y(1.01)

```
