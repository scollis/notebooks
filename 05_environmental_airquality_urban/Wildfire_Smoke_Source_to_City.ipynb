# Tracing Wildfire Smoke from Source to City: the Great Lakes & Northeast, July 2026

**Authors:** Scott Collis and Claude (Anthropic)

In July 2026, wildfire smoke from Canadian fires degraded air quality across the
US Midwest and Northeast, pushing cities from Chicago to Buffalo into the *Hazardous*
PM$_{2.5}$ range. This notebook traces the smoke from **source to surface impact** by
combining three open datasets:

- **NOAA Hazard Mapping System (HMS)** — daily satellite-analyzed *fire detections*
  and *smoke-plume polygons* (Light / Medium / Heavy density).
- **AirNow API** — real-time surface PM$_{2.5}$ from EPA / state / provincial monitors.
- **NOAA HRRR** — the High-Resolution Rapid Refresh model, from which we byte-range
  subset the **850 hPa wind field** to show the transport flow that steers the smoke.

We build a single figure: a regional map showing the Ontario fire source, the smoke
plumes draped across the Great Lakes, the HRRR 850 hPa wind field, and surface
PM$_{2.5}$ stations colored by EPA category — plus three time-series panels sampled
*along the plume's length* (Lake Superior → Chicago → western New York) to show the
smoke propagating downwind.

> **Reproducibility note.** The observation window, HMS date, and HRRR model run are
> hard-coded to the event (2026-07-16) so the notebook reproduces the same figure
> whenever it is run. The AirNow API key is read from the `AIRNOW_API` environment
> variable — set it before running: `export AIRNOW_API=your-key` (free key from
> https://docs.airnowapi.org/).


## 1. Setup and configuration

```python
import os
import json
import zipfile
import urllib.request
import urllib.parse
from collections import defaultdict
from datetime import datetime, timezone, timedelta
from zoneinfo import ZoneInfo

import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from scipy.interpolate import griddata

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
from shapely.geometry import box as sbox

import warnings
warnings.filterwarnings("ignore")

# --- Event configuration (pinned so the notebook reproduces this event) ---
HMS_DATE = "20260716"                 # NOAA HMS analysis date
END_DATE = "2026-07-16T17"            # end of the AirNow 24 h window (UTC hour)
WINDOW_HOURS = 24
CENTRAL = ZoneInfo("America/Chicago") # for local-time axes

# HRRR model run for the 850 hPa wind field (F00 analysis valid at END_DATE)
HRRR_RUN_DATE = "20260716"
HRRR_RUN_HOUR = "17"

# AirNow API key from environment (never hard-code credentials)
AIRNOW_API = os.environ["AIRNOW_API"]

# Regional extent: upper Midwest + Ontario source through the Northeast
# [lon_min, lon_max, lat_min, lat_max] -- east edge reaches New York State
REG = [-98, -72, 40, 52]

start_dt = datetime.strptime(END_DATE, "%Y-%m-%dT%H").replace(tzinfo=timezone.utc) - timedelta(hours=WINDOW_HOURS - 1)
START_DATE = start_dt.strftime("%Y-%m-%dT%H")
print("AirNow window:", START_DATE, "->", END_DATE, "UTC")
```

## 2. Download NOAA HMS fire detections and smoke plumes

HMS publishes daily shapefiles at `satepsanone.nesdis.noaa.gov`. We grab the
fire-point and smoke-polygon archives for the event date and unzip them locally.

```python
HMS_BASE = "https://satepsanone.nesdis.noaa.gov/pub/FIRE/web/HMS"
os.makedirs("hms", exist_ok=True)

hms_urls = {
    "smoke": f"{HMS_BASE}/Smoke_Polygons/Shapefile/{HMS_DATE[:4]}/{HMS_DATE[4:6]}/hms_smoke{HMS_DATE}.zip",
    "fire":  f"{HMS_BASE}/Fire_Points/Shapefile/{HMS_DATE[:4]}/{HMS_DATE[4:6]}/hms_fire{HMS_DATE}.zip",
}
for kind, url in hms_urls.items():
    dst = f"hms/{kind}.zip"
    if not os.path.exists(dst):
        urllib.request.urlretrieve(url, dst)
    with zipfile.ZipFile(dst) as z:
        z.extractall(f"hms/{kind}")
    print(kind, "->", os.path.getsize(dst), "bytes")

smoke_recs = list(shpreader.Reader(f"hms/smoke/hms_smoke{HMS_DATE}.shp").records())
fire_recs  = list(shpreader.Reader(f"hms/fire/hms_fire{HMS_DATE}.shp").records())
print(f"{len(smoke_recs)} smoke polygons, {len(fire_recs):,} fire detections (continent-wide)")
```

## 3. Quantify the source region

Restrict the fire detections to our regional window and split them roughly by
latitude to confirm the source is the Ontario fires (north of 48°N) rather than
the handful of detections in northern Minnesota.

```python
reg_box = sbox(REG[0], REG[2], REG[1], REG[3])

reg_fires = [(r.geometry.x, r.geometry.y) for r in fire_recs
             if REG[0] <= r.geometry.x <= REG[1] and REG[2] <= r.geometry.y <= REG[3]]
fx = np.array([p[0] for p in reg_fires])
fy = np.array([p[1] for p in reg_fires])

n_ontario = int((fy > 48.0).sum())
n_minn    = int(((fy <= 48.0) & (fx > -97) & (fx < -89)).sum())
print(f"Fire detections in region: {len(reg_fires):,}")
print(f"  north of 48N (Ontario / Canada): {n_ontario:,}")
print(f"  northern-Minnesota band:         {n_minn}")

for dens in ["Light", "Medium", "Heavy"]:
    n = sum(1 for r in smoke_recs
            if str(r.attributes.get("Density")) == dens and r.geometry.intersects(reg_box))
    print(f"  {dens:6s} smoke polygons over region: {n}")
```

## 4. Pull surface PM$_{2.5}$ from AirNow

A small helper wraps the AirNow `aq/data` endpoint. We pull the latest hour across
the whole region for the map, then 24 h series for three sites strung along the
plume axis.

```python
def airnow_data(bbox, start, end, param="PM25"):
    """Query the AirNow aq/data endpoint. bbox = 'minLon,minLat,maxLon,maxLat'."""
    q = urllib.parse.urlencode({
        "startDate": start, "endDate": end, "parameters": param, "BBOX": bbox,
        "dataType": "B", "format": "application/json",
        "verbose": 1, "monitorType": 2, "API_KEY": AIRNOW_API,
    })
    with urllib.request.urlopen("https://www.airnowapi.org/aq/data/?" + q, timeout=90) as r:
        return json.load(r)

# Latest-hour PM2.5 across the whole region (for the map)
reg_bbox = f"{REG[0]},{REG[2]},{REG[1]},{REG[3]}"
reg_raw = airnow_data(reg_bbox, END_DATE, END_DATE)
by_site = defaultdict(list)
for rec in reg_raw:
    if rec["Value"] is not None and rec["Value"] >= 0:
        by_site[(rec["Latitude"], rec["Longitude"], rec["SiteName"])].append(rec["Value"])
reg_stations = [(lat, lon, name, vals[-1]) for (lat, lon, name), vals in by_site.items()]
sv = np.array([s[3] for s in reg_stations])
print(f"{len(reg_stations)} AirNow stations | PM2.5 {sv.min():.0f}-{sv.max():.0f}, median {np.median(sv):.0f}")
```

```python
# Three sites along the plume axis: upstream (Lake Superior) -> mid (Chicago) ->
# downwind (western New York). Each entry is (bbox, agency_filter); the agency
# filter keeps the Buffalo pick a genuine New York State monitor (not an Ontario
# site that shares the border bbox).
PLUME_SITES = {
    "Duluth, MN":  ("-92.3,46.6,-91.9,47.0", None),
    "Chicago, IL": ("-88.1,41.6,-87.3,42.15", None),
    "Buffalo, NY": ("-79.0,42.8,-78.6,43.0", "New York"),
}

def best_series(bbox, start, end, agency_filter=None):
    """Return (site_name, (lat, lon), [(UTC, value), ...]) for the strongest site in bbox.
    Ranks by peak PM2.5 then completeness; agency_filter (substring of AgencyName)
    restricts to a reporting agency, e.g. "New York"."""
    raw = airnow_data(bbox, start, end)
    per, coords = defaultdict(list), {}
    for rec in raw:
        if rec["Value"] is not None and rec["Value"] >= 0:
            if agency_filter and agency_filter not in rec.get("AgencyName", ""):
                continue
            per[rec["SiteName"]].append((rec["UTC"], rec["Value"]))
            coords[rec["SiteName"]] = (rec["Latitude"], rec["Longitude"])
    site = max(per, key=lambda s: (max(v for _, v in per[s]), len(per[s])))
    return site, coords[site], sorted(per[site])

plume_series = {city: best_series(bbox, START_DATE, END_DATE, af)
                for city, (bbox, af) in PLUME_SITES.items()}
for city, (site, (la, lo), recs) in plume_series.items():
    arr = next((u for u, v in recs if v > 55), None)  # first hour above "Unhealthy"
    print(f"{city:12s} [{site:24s}] ({la:.2f},{lo:.2f}) peak {max(v for _,v in recs):5.0f}  "
          f"arrival>55: {arr[11:] if arr else '--'} UTC")
```

## 5. EPA PM$_{2.5}$ color scale

Standard EPA AQI breakpoints for 24 h PM$_{2.5}$, used both for the station colors
and the shaded bands behind the time series.

```python
EPA_BANDS = [
    (0.0,   12.0,  "#2c7fb8", "Good"),
    (12.0,  35.4,  "#fdae61", "Moderate"),
    (35.4,  55.4,  "#d7301f", "Unhealthy (SG)"),
    (55.4,  150.4, "#7a0177", "Unhealthy"),
    (150.4, 1e4,   "#4d004b", "Hazardous"),
]
def pm_color(v):
    for lo, hi, c, _ in EPA_BANDS:
        if v <= hi:
            return c
    return EPA_BANDS[-1][2]

SMOKE_COLORS = {"Light": "#ffe08a", "Medium": "#f59e3b", "Heavy": "#7a2b12"}
SMOKE_ALPHA  = {"Light": 0.12,      "Medium": 0.18,      "Heavy": 0.28}
```

## 6. HRRR 850 hPa wind field (byte-range GRIB2 subset)

To show the flow that steers the smoke, we add the **850 hPa wind** from NOAA's
High-Resolution Rapid Refresh (HRRR) model. HRRR pressure-level files are large
(~1 GB), so instead of downloading the whole file we read the `.idx` index, find the
byte offsets of just the `UGRD` and `VGRD` records at 850 mb, and request only that
byte range (~1 MB) from the NOAA Open-Data bucket on AWS. The F00 analysis from the
17Z run is valid at 2026-07-16 17:00 UTC — the same timestep as the AirNow and HMS
data, so the whole figure is one synoptic snapshot.

```python
# NOAA HRRR on AWS Open Data: byte-range subset of just 850 mb U/V winds.
HRRR_BASE = "https://noaa-hrrr-bdp-pds.s3.amazonaws.com"
grib_url = f"{HRRR_BASE}/hrrr.{HRRR_RUN_DATE}/conus/hrrr.t{HRRR_RUN_HOUR}z.wrfprsf00.grib2"
idx_url = grib_url + ".idx"

# 1) read the index, locate the 850 mb UGRD/VGRD records and their byte offsets
idx_lines = urllib.request.urlopen(idx_url, timeout=60).read().decode().splitlines()
starts = {int(l.split(":")[0]): int(l.split(":")[1]) for l in idx_lines}
want = {}
for l in idx_lines:
    p = l.split(":")
    if p[4] == "850 mb" and p[3] in ("UGRD", "VGRD"):
        want[p[3]] = int(p[0])
n_u, n_v = want["UGRD"], want["VGRD"]

# 2) download the contiguous byte range spanning both records
b0 = starts[n_u]
b1 = starts.get(max(n_u, n_v) + 1)
rng = f"bytes={b0}-" + (str(b1 - 1) if b1 else "")
req = urllib.request.Request(grib_url, headers={"Range": rng})
with urllib.request.urlopen(req, timeout=120) as r:
    grib_bytes = r.read()
with open("hrrr_850_uv.grib2", "wb") as f:
    f.write(grib_bytes)
print(f"HRRR 850 mb U/V: {len(grib_bytes)/1e6:.2f} MB (records {n_u}, {n_v})")

# 3) read with cfgrib and subset to the map region (+1 deg margin)
ds_hrrr = xr.open_dataset("hrrr_850_uv.grib2", engine="cfgrib",
                          backend_kwargs={"indexpath": ""})
hlat = ds_hrrr["latitude"].values
hlon = ds_hrrr["longitude"].values
hlon = np.where(hlon > 180, hlon - 360, hlon)
hu, hv = ds_hrrr["u"].values, ds_hrrr["v"].values
m = (hlat >= REG[2]-1) & (hlat <= REG[3]+1) & (hlon >= REG[0]-1) & (hlon <= REG[1]+1)
ys, xs = np.where(m)
y0, y1, x0, x1 = ys.min(), ys.max()+1, xs.min(), xs.max()+1
sub_lat = hlat[y0:y1, x0:x1]; sub_lon = hlon[y0:y1, x0:x1]
sub_u = hu[y0:y1, x0:x1];     sub_v = hv[y0:y1, x0:x1]
valid = str(ds_hrrr["valid_time"].values)[:16]
corr = (sub_lat >= 43) & (sub_lat <= 50) & (sub_lon >= -92) & (sub_lon <= -75)
print(f"HRRR valid {valid} | subgrid {sub_lat.shape} | "
      f"corridor mean U,V = {sub_u[corr].mean():.1f}, {sub_v[corr].mean():.1f} m/s")
```

## 7. The composite figure

```python
proj = ccrs.LambertConformal(central_longitude=-85, central_latitude=46)
fig = plt.figure(figsize=(13.5, 12.5))
gs = fig.add_gridspec(2, 3, height_ratios=[2.5, 1.0], hspace=0.17, wspace=0.14,
                      left=0.045, right=0.97, top=0.945, bottom=0.075)

# ---- Map ----
ax = fig.add_subplot(gs[0, :], projection=proj)
ax.set_extent(REG, crs=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND.with_scale("50m"),  facecolor="#f3efe6", zorder=0)
ax.add_feature(cfeature.OCEAN.with_scale("50m"), facecolor="#d7e6ef", zorder=0)
ax.add_feature(cfeature.LAKES.with_scale("50m"), facecolor="#d7e6ef",
               edgecolor="#9fb8c8", linewidth=0.5, zorder=1)
ax.add_feature(cfeature.STATES.with_scale("50m"),  edgecolor="#9a9a9a", linewidth=0.5, zorder=2)
ax.add_feature(cfeature.BORDERS.with_scale("50m"), edgecolor="#333333", linewidth=1.1, zorder=2.5)
ax.coastlines("50m", color="#6a6a6a", linewidth=0.6, zorder=2)
pc = ccrs.PlateCarree()

# smoke plumes + fire detections
for dens in ["Light", "Medium", "Heavy"]:
    geoms = [r.geometry for r in smoke_recs if str(r.attributes.get("Density")) == dens]
    ax.add_geometries(geoms, pc, facecolor=SMOKE_COLORS[dens],
                      edgecolor=SMOKE_COLORS[dens], linewidth=0.6,
                      alpha=SMOKE_ALPHA[dens], zorder=3)
ax.scatter(fx, fy, s=5, c="#c81e1e", marker="o", transform=pc, zorder=4, alpha=0.28, linewidths=0)

# HRRR 850 hPa wind: gray streamlines (shaded by speed) + subsampled barbs (knots)
KT = 1.94384
gx = np.linspace(REG[0], REG[1], 60); gy = np.linspace(REG[2], REG[3], 34)
GX, GY = np.meshgrid(gx, gy)
pts = np.column_stack([sub_lon.ravel(), sub_lat.ravel()])
GU = griddata(pts, sub_u.ravel(), (GX, GY), method="linear")
GV = griddata(pts, sub_v.ravel(), (GX, GY), method="linear")
GS = np.sqrt(GU**2 + GV**2)
ax.streamplot(gx, gy, GU, GV, transform=pc, color=GS, cmap="Greys",
              density=1.5, linewidth=0.7, arrowsize=0.8, zorder=5.5,
              norm=plt.Normalize(0, 18))
step = 28
ax.barbs(sub_lon[::step, ::step], sub_lat[::step, ::step],
         sub_u[::step, ::step]*KT, sub_v[::step, ::step]*KT,
         length=5.2, transform=pc, zorder=6, color="#1a1a1a", linewidth=0.55,
         barb_increments=dict(half=5, full=10, flag=50))

# surface PM2.5 stations
for lat, lon, name, v in reg_stations:
    ax.scatter(lon, lat, s=20 + 6*np.sqrt(v), c=pm_color(v), edgecolor="white",
               linewidth=0.35, transform=pc, zorder=7, alpha=0.92)
ax.text(-88.5, 51.0, "Ontario wildfires\n(source region)", transform=pc,
        fontsize=10.5, fontweight="bold", color="#7a2b12", zorder=9, ha="center", va="center")

tags = ["a", "b", "c"]
for tag, (city, (site, (la, lo), recs)) in zip(tags, plume_series.items()):
    ax.scatter(lo, la, s=300, marker="*", facecolor="#111111", edgecolor="white",
               linewidth=1.2, transform=pc, zorder=8)
    ax.text(lo + 0.4, la + 0.3, f"({tag}) {city.split(',')[0]}", transform=pc,
            fontsize=9.5, fontweight="bold", zorder=9, ha="left", va="bottom",
            bbox=dict(boxstyle="round,pad=0.18", fc="white", ec="#111", lw=0.5, alpha=0.9))

src_handles = [Line2D([], [], marker="o", color="#c81e1e", ls="none", ms=6, alpha=0.7,
                      label=f"Fire detections (n={len(reg_fires):,})")] + \
              [Patch(facecolor=SMOKE_COLORS[d], alpha=min(SMOKE_ALPHA[d] + 0.3, 0.7),
                     label=f"{d} smoke") for d in ["Light", "Medium", "Heavy"]] + \
              [Line2D([], [], marker=r"$\rightarrow$", color="#1a1a1a", ls="none", ms=11,
                      label="850 hPa wind (barbs/flow)")]
leg1 = ax.legend(handles=src_handles, loc="lower left", fontsize=8.2, framealpha=0.93,
                 title=f"Source & flow - NOAA HMS {HMS_DATE} + HRRR 850 hPa", title_fontsize=8.2)
leg1.set_zorder(12); ax.add_artist(leg1)
epa_handles = [Patch(facecolor=c, edgecolor="white", label=lbl) for _, _, c, lbl in EPA_BANDS]
ax.legend(handles=epa_handles, loc="lower right", fontsize=8.2, framealpha=0.93,
          title=f"Impact - AirNow PM2.5 ({END_DATE}Z)", title_fontsize=8.2).set_zorder(12)
ax.set_title("Wildfire smoke Ontario -> Northeast with HRRR 850 hPa flow - source, wind, and surface impact\n"
             "NOAA HMS fires + smoke plumes, HRRR 850 hPa wind, AirNow surface PM$_{2.5}$ (valid 2026-07-16 17Z)",
             fontsize=12.5, fontweight="bold")

# ---- Three time series along the plume (west -> east) ----
ymax = max(max(v for _, v in recs) for _, (_, _, recs) in plume_series.items()) * 1.10
for i, (tag, (city, (site, _, recs))) in enumerate(zip(tags, plume_series.items())):
    axb = fig.add_subplot(gs[1, i])
    t = [datetime.strptime(u, "%Y-%m-%dT%H:%M").replace(tzinfo=timezone.utc).astimezone(CENTRAL)
         for u, _ in recs]
    y = [v for _, v in recs]
    for lo, hi, col, _ in EPA_BANDS:
        axb.axhspan(lo, min(hi, ymax), color=col, alpha=0.13, zorder=0, linewidth=0)
    axb.plot(t, y, "-o", color="#222222", ms=3.2, lw=1.4, zorder=3)
    axb.set_ylim(0, ymax); axb.set_xlim(t[0], t[-1])
    axb.set_title(f"({tag}) {city} - {site}", fontsize=9.5, fontweight="bold")
    axb.xaxis.set_major_formatter(mdates.DateFormatter("%-I%p", tz=CENTRAL))
    axb.xaxis.set_major_locator(mdates.HourLocator(interval=6, tz=CENTRAL))
    axb.set_xlabel("Local time (CDT)", fontsize=9)
    if i == 0:
        axb.set_ylabel("PM$_{2.5}$  (µg/m³)", fontsize=9.5)
    else:
        axb.tick_params(labelleft=False)
    axb.tick_params(labelsize=8.5)

fig.savefig("smoke_source_impact_wind_ts.png", dpi=130)
plt.show()
```
